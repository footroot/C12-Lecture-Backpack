# Perceptron

This notebook implements a neural network from scratch to gain an understanding of how neural networks work, and that is essential for designing effective models. Here, we create a perceptron i.e. a neural network with an input and output layer.

```python
import numpy as np
```

### Activation Functions

Sigmoid activation function for neural networks

```python
def sigmoid(x):
    # This function maps any value to a value between 0 and 1, which can be interpreted as a probability.
    return 1 / (1 + np.exp(-x))
```

Tanh activation function

```python
def tanh(x):
    # This function maps any value to a value between -1 and 1.
    y=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    return y
```

Plot the Sigmoid and Tanh activation functions

```python
import matplotlib.pyplot as plt 
z=np.arange(-4,4,0.01)

# Setup centered axes
fig, ax = plt.subplots(figsize=(7, 4))
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Plot both the activation functions
ax.plot(z,tanh(z), linewidth=3, label="Tanh")
ax.plot(z,sigmoid(z), linewidth=3, label="Sigmoid")
ax.legend()
```

ReLU activation function

```python
def relu(x):
    return max(0,x)

print(relu(-10))
print(relu(7))
```

Leaky ReLU

```python
def leaky_relu(x, alpha=0.1):
    return max(x, alpha * x)

print(leaky_relu(-10))
print(leaky_relu(7))
```

Plot ReLU and Leaky ReLU

```python
# Generate values for x
x = np.linspace(-10, 10, 100)

# Compute ReLU and Leaky ReLU values for corresponding x
relu = np.maximum(0, x)
alpha = 0.1
leaky_relu_values = np.where(x >= 0, x, alpha * x)

# Plot the ReLU function
plt.plot(x, relu,label='ReLU Activation Function',lw=3)
plt.plot(x, leaky_relu_values,label='Leaky ReLU Activation Function',lw=3)
plt.xlabel("x")
plt.ylabel("Activation Function")
plt.grid(True)
plt.legend()
```

### Input and Output layers of the Neural Network

Define a single layer of a neural network. Here we perform matrix multiplication between inputs and weights and then add the biases. This linear algebra operation is the basis of data transformation in neural networks. 

We the apply the sigmoid activation function to introduce non-linearity. Non-linearity allows the network to learn more complex patterns.

```python
def neural_network_layer(inputs, weights, biases):
    layer_input = np.dot(inputs, weights) + biases
    #Apply the sigmoid activation function
    layer_output = sigmoid(layer_input)
    return layer_output
```

Next, we initialise the weights and biases for the input layer. 

Here, let us consider that there are two input nodes and one output node. Two weights are required hence, and one bias.

```python
input_node = 2
output_node = 1
```

Initialise weights (for 2 inputs to 1 output) and biases for the layer

```python
# We can set the weights and biases as input like below
#weights = np.array([[0.2], [-0.3]])
#biases = np.array([0.1])

# Usually weights are randomly choosen, and biases can get set to zero. 
weights = np.random.randn(input_node,output_node)
#biases = np.zeros((1))

#However, let us also use a random value for the bias. 
biases = np.random.randn((1))

print("The weights are: ", weights)
print("The bias is: ", biases)
```

```python
# Example input vector (2 inputs) or we can have an user input
inputs = np.array([0.1, 0.25])
```

```python
# Perform a forward pass through the network layer
output = neural_network_layer(inputs, weights, biases)
print("Neural Network Output:", output)
```
