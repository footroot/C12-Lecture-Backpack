Imports

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

Load the sales dataset

```python
sales_df = pd.read_csv("retail_sales_dataset.csv")
```

Data Manipulation: Filter sales based on transaction value

```python
high_sales = sales_df[sales_df["Total Amount"] > 1000]
high_sales
```

Total revenue per category: pandas

```python
total_p = sales_df.groupby("Product Category")["Total Amount"].sum()
total_p 
```

Total per category: Iteration

```python
beauty = sales_df[sales_df["Product Category"] == "Beauty"]["Total Amount"].values
clothing = sales_df[sales_df["Product Category"] == "Clothing"]["Total Amount"].values
elec = sales_df[sales_df["Product Category"] == "Electronics"]["Total Amount"].values

total_b = 0
total_c = 0
total_e = 0

for b in beauty:
    total_b += b

for c in clothing:
    total_c += c

for e in elec:
    total_e += e

print("Beauty: $", total_b)
print("Clothing: $", total_c)
print("Electronics: $", total_e)
```

Create sales threshold column

Create a function to label each total amount

```python
def label_sales(row):
    total = row["Total Amount"]

    if total > 2000:
        label = "High"
    elif total > 800:
        label = "Medium"
    else: 
        label = "Low"
    
    return label
```

```python
sales_df["Sales Threshold"] = sales_df.apply(label_sales, axis=1)
sales_df
```
