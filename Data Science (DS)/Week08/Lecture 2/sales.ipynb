```python
import pandas as pd

df = pd.read_csv('sales_data.csv')

df.head()
```

```python
import matplotlib.pyplot as plt

category_sales = df.groupby('Category')['Revenue'].sum()

category_sales.head()
```

```python
# plot
category_sales.plot(kind='bar', color='skyblue')
plt.title("Total sales by category")
plt.xlabel("Category")
plt.ylabel("Revenue ($)")
plt.xticks(rotation=45)
plt.show()
```

```python
# pie chart
region_sales = df.groupby('Region')['Revenue'].sum()

plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightblue', 'pink', 'green'])
plt.title("Revenue Distribution by Region")
plt.show()
```

```python
# scatterplot

plt.scatter(df['Ad Spend'], df['Revenue'], color="purple", alpha=0.7)
plt.title("Ad Spend vs Revenue")
plt.xlabel('Ad Spend ($)')
plt.ylabel('Revenue')
plt.show()
```

```python
# line graph
monthly_sales = df.groupby('Month')['Revenue'].sum()

plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='blue')
plt.title("Monthly Revenue Trends")
plt.xlabel('Month')
plt.ylabel('Revenue ($)')
plt.show()
```

```python
# boxplot
import seaborn as sns

sns.boxplot(data=df, x="Category", y='Revenue')
plt.title("Revenue Distribution by Category")
plt.show()
```

```python
# pairplot
sns.pairplot(df, hue="Category")
plt.show()
```

```python
corr_matrix = df[['Revenue', 'Ad Spend', 'Units Sold', 'Customer Rating']].corr()

sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()
```
